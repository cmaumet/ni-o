## DICOM Term Extraction

#### Retrival of DICOM terms from NeuroLex

This notebook is a tutorial on extracting terms from http://neurolex.org, and, more specifically, DICOM terms.

Neurolex is being used as a location to:

* register terms
* provide a URI
* provide a human readable URL
* provide definitions

The data was downloaded from Neurolex (http://neurolex.org/OWL/nlx_stage_all.rdf) then uploaded to a virtuoso triplestore.

Note: The SPARQL endpoint at http://rdf-stage.neuinfo.org/sparql isn't working as of 06/13/2014

```python
import StringIO
import pandas as pd
from rdflib import ConjunctiveGraph
```

```python
g = ConjunctiveGraph(store='SPARQLStore')
g.open("http://localhost:8890/sparql")
#g.open("http://rdf-stage.neuinfo.org/sparql")
```

```python
select_query = """
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX property: <http://neurolex.org/wiki/Property-3A>

SELECT DISTINCT ?label ?termURL ?sourceIRI ?dicomTag ?definition ?vr
WHERE {?termURL property:SuperCategory "DICOM term"^^xsd:string .
       ?termURL property:Definition ?definition .
       ?termURL property:DICOMID ?dicomTag .
       ?termURL property:Id ?nlxID .
       ?termURL rdfs:isDefinedBy ?sourceIRI .
       ?termURL rdfs:label ?label .
       ?termURL property:Value_Representation ?vr .}
"""
```

```python
select_results = g.store.query(select_query)
```

```python
construct_query = """
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX nlx: <http://neurolex.org/wiki/>
PREFIX property: <http://neurolex.org/wiki/Property-3A>
PREFIX obo: <http://purl.obolibrary.org/obo/>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX dcm: <http://www.incf.org/ns/nidash/dicom#>

CONSTRUCT {?sourceIRI a owl:DatatypeProperty ;
                      rdfs:label ?label ;
                      rdfs:seeAlso ?termURL ;
                      property:DICOMID ?dicomTag ; 
                      obo:IAO_0000115 ?definition .
           }
WHERE {?termURL property:SuperCategory "DICOM term"^^xsd:string ;
            property:Definition ?definition ;
            property:DICOMID ?dicomTag ;
            property:Id ?nlxID ;
            rdfs:isDefinedBy ?sourceIRI ;
            rdfs:label ?label ;
            property:Value_Representation ?vr .
        FILTER (!regex(str(?sourceIRI), 'http://neurolex.org/wiki/Special:ExportRDF/Nlx', 'i'))
        }
"""
```

```python
construct_results = g.store.query(construct_query)
```
